```python
import pandas as pd
import mysql.connector

# Load CSV into DataFrame
df=pd.read_csv('course_data.csv')
# Connect to MySQL
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Paperkraft@#41',
    database='course_data'
)

# Create a MySQL cursor
cursor = connection.cursor()

# Execute queries to create a new database and table
cursor.execute("CREATE DATABASE IF NOT EXISTS course_data")
cursor.execute("USE course_data")
cursor.execute("""
    CREATE TABLE IF NOT EXISTS course_table (
        id INT AUTO_INCREMENT PRIMARY KEY,
        course_name VARCHAR(255),
        course_discipline VARCHAR(255),
        link_to_course VARCHAR(255)
    )
""")


```

```python
# Load CSV Data into Table
for index, row in df.iterrows():
    cursor.execute("""
        INSERT INTO course_table (course_name, course_discipline, link_to_course)
        VALUES (%s, %s, %s)
    """, (row['name'], row['discipline'], row['link']))

# Verify Data
cursor.execute("SELECT * FROM course_table")
result = cursor.fetchall()
for row in result:
    print(row)

# Commit changes and close connection
connection.commit()
connection.close()
```
